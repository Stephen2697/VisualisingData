# Relationship plots

This week we want to explore our data to see what the correlation is between the different variables.  To do this, we'll use a range of chart types, sometimes using ggplot.

```r
library(tidyverse)
library(reshape2)
library(sqldf)
```

scatterplot
line column charts with a secondary axis and 
bubble plot
SPLOM
heat map

```r
data('mtcars')
```

```r
str(mtcars)
```

```r
options(repr.plot.width=16,repr.plot.height=16)
pairs(mtcars)
```

```r
df=sqldf("select mpg, drat, wt, qsec from mtcars")
```

```r
str(df)
```

```r
options(repr.plot.width=8,repr.plot.height=8)
pairs(df)
```

```r
library("GGally")
```

```r
ggpairs(df)
```

```r
ggcorr(df, palette = "RdBu", label = TRUE)
```

```r
mtcars$cyl=as.factor(mtcars$cyl)
```

```r
p = ggplot(mtcars, aes(x = wt, y = mpg)) + 
  geom_point(aes(color = cyl, size = qsec), alpha = 0.5) +
  scale_color_manual(values = c("#00AFBB", "#E7B800", "#FC4E07")) +
  scale_size(range = c(1, 12))  # Adjust the range of points size
p
```
