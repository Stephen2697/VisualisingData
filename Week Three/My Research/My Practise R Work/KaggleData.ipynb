# House Data Visualisation in R

#### Read In Sample Data: https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data


```r
data <- read.csv("Input.csv")
```

```r
HouseData = subset( data, select = c(Id, LotFrontage, LotArea, Neighborhood, OverallQual, RoofStyle, YearBuilt, YearRemodAdd, MoSold, YrSold,SaleType, SaleCondition, SalePrice, BedroomAbvGr) )
```

#### Show a summary of your dataset using str.

```r
str(HouseData)
```

```r
head(HouseData,5)
```

```r
summary(HouseData)
```

#### Get the average house price.

```r
mean(HouseData$SalePrice, na.rm = TRUE) 
```

#### Using the plot() command, plot all variables against each other. Take note of any interesting correlations and do further plots on those pairs.

```r
options(repr.plot.width=16, repr.plot.height=16)
plot(HouseData)
```

```r
options(repr.plot.width=16, repr.plot.height=16)
plot(HouseData$SaleCondition, HouseData$SalePrice, main = 'Correlation: Sale Condition to Sale Price',xlab = 'House Condition', ylab = 'Sale Price')
```

```r
plot(data$YrSold,data$SalePrice,type="h")
```

```r
options(repr.plot.width=16, repr.plot.height=16)
plot(HouseData$BedroomAbvGr,HouseData$SalePrice,main = 'Correlation: Number of Bedrooms to Sale Price',xlab = 'Bedrooms Above Ground', ylab = 'Sale Price')
```

```r
options(repr.plot.width=16, repr.plot.height=16)
plot(data$YearBuilt,data$SalePrice,type="h")
```

#### Plot the dimensions OverallQual, SalePrice, Neighborhood and BedroomAbvGR on the same chart:


```r
library(ggplot2)
options(repr.plot.width=16, repr.plot.height=16)
plot <- qplot(data$OverallQual , data$BedroomAbvGr , size = data$SalePrice , color = data$Neighborhood , main="Graph")+theme(legend.position = 'right')

print(plot)
```

#### Using the boxplot function or barplot function, show the distribution over the variables: LotFrontage, LotArea, Neighborhood, OverallQual, RoofStyle, YearBuilt, YearRemodAdd, YrSold, SaleType, SaleCondition, SalePrice, BedroomAbvGr showing them all in a 3x3 square of charts.

```r
options(repr.plot.width=16, repr.plot.height=16)
par(mfrow=c(4,3), mar=c(2,5,2,1), las=1, bty="n")

plot(data$Neighborhood,main='Neighbourhood',col='green')

boxplot(data$YearBuilt,main='Year Built')

barplot(data$SalePrice, main = 'SalePrice',xlab = 'SalePrice',horiz = FALSE)

plot(data$BedroomAbvGr,main='Number of Bedrooms')

barplot(data$LotArea, main = 'Lot Area',xlab = 'Lot Area',horiz = FALSE)

plot(data$YrSold,main='Year Sold')

boxplot(data$LotArea,main='Lot Area')

boxplot(data$YearRemodAdd,main='Year Remodeled')

barplot(data$OverallQual, main = 'Overall Quality',xlab = 'Overall Quality',horiz = TRUE)

plot(data$Neighborhood,data$OverallQual,xlab = 'Neighbourhood' , ylab = "Overall Quality",type="h")

plot(data$RoofStyle, main = 'Roof Style',xlab = 'Roof Style', col='blue',horiz = FALSE)

boxplot(data[,0:4], main='Combined Box plots')









```
