## **[Step One: Package Installation]**

```r
#Do Installation:
install.packages("tidyverse") #includes ggplot2 & many useful module extensions
install.packages("cowplot")

#Import Libraries:
library(tidyverse)
library(cowplot)

#Set Environment Plot Size:
options(repr.plot.width=15, repr.plot.height=10)

#sessionInfo() 
#Check all installed datasets we can use:
#data()

#Define my Own Theme                           
my_Theme <- theme(plot.title = element_text(face="bold",size=20),
                  plot.subtitle = element_text(face="italic", size = 12),
                  axis.title = element_text(face="bold",size=20),
                  axis.text = element_text(size=20),
                  plot.caption = element_text(size = 14))
                           
```

---

## **[Step Two: Dataset Found & Used]**

```r
#Load Example Data:
data("airquality")

#Check the Documentation & Grab a Quick Visual of the Data:
#?airquality
plot(airquality)
glimpse(airquality)
```

---

## **[Step Three: Histogram & Frequency Polygon Plot]**

```r
histPlot <- airquality %>% 
    ggplot(aes(Temp)) + 
    geom_histogram(aes(fill=..count..), bins=31, show.legend = F) +
    scale_fill_gradient("Count", low="cyan", high="purple")+ 
    theme_bw()+my_Theme+
    labs(x = "Temperature [°F]", y="Frequency Count",
        title="Temperature Reading Frequency Histogram",
        subtitle="Analysis for the Month of May, Maximum daily temperature in degrees\nFahrenheit at La Guardia Airport. Temperatures in the 75-85 Degree Fahrenheit\nRange were observed with High Frequency.",
        caption="R Dataset Source: [airquality] © Stephen Alger 2021") 
```

```r
frePlot <- airquality %>% 
    ggplot(aes(Wind)) + 
    geom_freqpoly(size=5, colour="red", bins=31)+ 
    theme_bw()+my_Theme+
    labs(x = "Wind [MPH]", y="Frequency Count",
        title="Windage Reading Frequency Polygon",
        subtitle="Analysis for the Month of May, Average wind speed in miles per hour\nat 0700 and 1000 hours at LaGuardia Airport. Windage Readings in the 8-12 Miles\nPer Hour Range were observed with High Frequency.",
        caption="R Dataset Source: [airquality] © Stephen Alger 2021")
```

```r
plot_grid(histPlot, frePlot, labels = "AUTO")
```

---

## **[Step Four: Boxplot For Three Columns in dataframe]**

```r
boxPlot <- airquality %>%
    ggplot(aes(x= factor(Day), y=Temp)) + 
    geom_boxplot()+
    geom_point(alpha = 0.8,size=2, aes(colour = Temp))+
    scale_color_gradient(low="yellow", high="red")+
    theme_bw()+my_Theme+
    labs(x = "Date in May", y="Temperature [°F]",
        title="Temperature Box Plot",
        subtitle="Analysis for the Month of May, Maximum daily temperature in degrees Fahrenheit at La Guardia Airport Per Day.",
        caption="R Dataset Source: [airquality] © Stephen Alger 2021")
    
```

```r
boxPlot2 <- airquality %>%
    ggplot(aes(x= factor(Day), y=Wind)) + 
    geom_boxplot()+
    geom_point(alpha = 0.8,size=2, aes(colour = Wind))+
    scale_color_gradient(low="purple", high="cyan")+
    theme_bw()+my_Theme+
    labs(x = "Date in May", y="Wind Speed [MPH]",
        title="Windage Box Plot",
        subtitle="Analysis for the Month of May, Average wind speed in miles per hour at 0700 and 1000 hours at LaGuardia Airport Per Day.",
        caption="R Dataset Source: [airquality] © Stephen Alger 2021")
```

```r
boxPlot3 <- airquality %>%
    ggplot(aes(x= factor(Day), y=Ozone)) + 
    geom_boxplot()+
    geom_point(alpha = 0.8,size=2, aes(colour = Ozone))+
    scale_color_gradient(low="green", high="cyan")+
    theme_bw()+my_Theme+
    labs(x = "Date in May", y="Ozone [bpb]",
        title="Ozone Parts Per Billion Box Plot",
        subtitle="Analysis for the Month of May, Mean ozone in parts per billion from 1300 to 1500 hours at Roosevelt Island Per Day.",
        caption="R Dataset Source: [airquality] © Stephen Alger 2021")
```

```r
plot_grid(boxPlot, boxPlot2, boxPlot3, labels = "AUTO")
```

## **[Step Five: Bar Graph For Single Variable in dataframe]**

```r
Bar <- airquality %>%
    ggplot(aes(Wind))+
    geom_bar(aes(fill=..count..),width=0.1, show.legend = F)+theme_bw()+my_Theme+
    scale_fill_gradient("Count", low="cyan", high="purple")+ 
    labs(x = "Wind [MPH]", y="Frequency Count",
        title="Windage Reading Frequency Barchart",
        subtitle="Analysis for the Month of May, Average wind speed in miles per hour\nat 0700 and 1000 hours at LaGuardia Airport. Windage Readings in the 8-12 Miles\nPer Hour Range were observed with High Frequency.",
        caption="R Dataset Source: [airquality] © Stephen Alger 2021")
Bar        
  
```

## [**Step Six: Bar Graph Representing Two Variables]**


```r
#Pretty Terrible dataset for this exercise so I am simply showing how this works regardless of the lack of meaning in the chart.
#A pair of overlayed frequency barchcharts, X is the Temperature frequency distribution with a Wind frequency barchart overlay

dualBar <- airquality %>%
    ggplot(aes(x=factor(Temp), fill=factor(Wind))) + 
    geom_bar() +
    labs(x = "Temperature & Wind", y="Frequency",
        title="Two Variables Overlayed In One Barchart",
        subtitle="Not a particularly enlightening chart, Illustration of Implementation.\nWould have used a different dataset but had instructions to use one & insufficient time to familiarise myself with another",
        caption="R Dataset Source: [airquality] © Stephen Alger 2021")+
    theme_bw()+my_Theme
dualBar
```

 ## [**Step Seven: pie chart representing one of the variables from your dataframe]**


```r
#Again: Terrible dataset for this exercise so I am simply showing how this works regardless of the lack of meaning in the chart.

dataframe <- as.data.frame(table(airquality$Temp))
colnames(dataframe) <- c("class", "frequency")

pie <- ggplot(dataframe, aes(x = "", y=frequency, fill = factor(class))) +
    geom_bar(stat="identity", colour="black")+
    labs(x = NULL, y=NULL,
        title="Piechart of Temperature Frequency",
        subtitle="Not a particularly enlightening chart as Temperatures in themselves are\nnot categorical data per se - This dataset is not particularly suited for a piechart",
        caption="R Dataset Source: [airquality] © Stephen Alger 2021")+
    coord_polar(theta = "y", start=0)+
    theme_bw()+my_Theme
pie
```

 ## [**Step Eight: Rose chart representing one of the variables from your dataframe]**

```r
narrowRosePlot <- ggplot(data = airquality) +
    geom_bar(aes(x = Wind, fill = factor(Wind)),show.legend = F,width = 0.1) +
    coord_polar()+
    labs(x = NULL, y=NULL,
        title="Narrow Slice Rose Chart of Wind Frequency",
        subtitle="Not a particularly enlightening chart as Wind speeds in themselves are no\ncategorical data per se - This dataset is not particularly suited for a piechart",
        caption="R Dataset Source: [airquality] © Stephen Alger 2021")+
    theme_bw()+my_Theme+
    theme(axis.ticks=element_blank(),
          axis.text.y=element_blank(),
          axis.title=element_blank()) 
```

```r
wideRosePlot <- ggplot(data = airquality) +
    geom_bar(aes(x = Wind, fill = factor(Wind)),colour="black",show.legend = F,width = 1) +
    coord_polar()+
    labs(x = NULL, y=NULL,
        title="Wide Slice Rose Chart of Wind Frequency",
        subtitle="Not a particularly enlightening chart as Wind speeds in themselves are not\ncategorical data per se - This dataset is not particularly suited for a piechart",
        caption="R Dataset Source: [airquality] © Stephen Alger 2021")+
    theme_bw()+my_Theme+
    theme(axis.ticks=element_blank(),
          axis.text.y=element_blank(),
          axis.title=element_blank()) 
```

```r
plot_grid(wideRosePlot, narrowRosePlot, labels = "AUTO")
```

---

 ## [**Step Nine: Wind Rose Pie Chart]**


```r
rosePiePlot <- 
    airquality %>%
    ggplot(aes(x = factor(Wind), y = factor(Temp), fill = factor(Wind))) + 
        geom_bar(width = 2, stat="identity",colour="black") + coord_polar("y", start=pi / 3) + 
        ggtitle("Wind Rose Pie Chart") + 
        labs(x = NULL, y=NULL,
             title="Wind Rose Pie Chart of Wind & Temperature Frequency",
            subtitle="Not a particularly enlightening chart as Wind speed overlayed on Temperatures speed piechart - This dataset is not particularly suited for a piechart",
            caption="R Dataset Source: [airquality] © Stephen Alger 2021")+
        theme_bw()+
        theme(plot.title = element_text(face="bold",size=20),
            plot.subtitle = element_text(face="italic", size = 12),
            plot.caption = element_text(size = 14))

rosePiePlot
```

```r
dualStackPie <- ggplot(data = airquality) +
    geom_bar(aes(x = factor(Temp), fill = factor(Wind)),colour="black",show.legend = F,width = 1) +
    coord_polar()+
    labs(x = NULL, y=NULL,
        title="Dual Variable Rose Chart of Wind Frequency",
        subtitle="Not a particularly enlightening chart as Wind speed overlayed on\nTemperature speed piechart - Just Exploring Visualisation",
        caption="R Dataset Source: [airquality] © Stephen Alger 2021")+
    theme_bw()+my_Theme+
    theme(axis.ticks=element_blank(),
          axis.text.y=element_blank(),
          axis.title=element_blank()) 

dualStackPie
```
