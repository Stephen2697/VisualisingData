 - The data is what needs to be plotted.
 - Graphics are made up of distinct layers of grammatical elements.
 - Meaningful plots are built around appropriate aesthetic mappings. 
 - The aesthetics layer defines the scales where the mapping takes place.
 - The geom layer refers to the shape the data will take.

## Histogram and frequency polygon

```r
ggplot(data=mtcars, aes(mpg)) + 
  geom_histogram()
```

 The default number of bins is 30.  Let's change it to 15.

```r
ggplot(data=mtcars, aes(mpg)) + 
  geom_histogram(bins=15)
```

I can use a frequency polygon instead:

```r
ggplot(data=mtcars, aes(mpg)) + 
  geom_freqpoly(bins=15)
```

#### Combine the histogram and frequency polygon

```r
ggplot(data=mtcars, aes(mpg)) + 
  geom_histogram(bins=15) +
  geom_freqpoly(bins=15)
```

Add some colours

```r
ggplot(data=mtcars, aes(mpg)) + 
  geom_histogram(bins=15, fill="orange") +
  geom_freqpoly(bins=15,colour="blue", size=1)
```

## Box plot

A box plot shows the distribution of values in the column.  It doesn't make much sense to do a box plot of the cylinders, but one of the mpg makes a bit more sense

```r
ggplot(data=mtcars, aes(cyl)) + 
geom_boxplot()
```

```r
ggplot(data=mtcars, aes(mpg)) + 
geom_boxplot()
```

```r
ggplot(data=mtcars, aes(drat)) + 
geom_boxplot()
```

### Compare these results with the histogram

```r
ggplot(data=mtcars, aes(drat)) + 
  geom_histogram(bins=15, color="blue", fill="yellow")
```

## Bar graph
These show the number of rows for each value of the column

```r
ggplot(mtcars, aes(cyl)) + geom_bar()
```

Note:  We can save our plot layers and build on them

```r
p<-ggplot(mtcars, aes(cyl)) 
p<-p+geom_bar()
p
```

Because there is a column for every value, this doesn't work well for real numbers

```r
p<-ggplot(mtcars, aes(mpg))
p+geom_bar()
```

We could make it categorical, by using the as.factor function

```r
df<-mtcars
df$mpg<-as.factor(df$mpg)
p<-ggplot(df, aes(mpg))
p+geom_bar()
```

Or integer, by using the as.integer function

```r
df<-mtcars
df$mpg<-as.integer(df$mpg)
p<-ggplot(df, aes(mpg))
p+geom_bar()
```

Let's try combining our gears and cylinder information, so we can see e.g. the number of 3-cylinder cars that have 4 gears:

```r
ggplot(data = mtcars, aes(x=gear, y=cyl, fill=cyl) ) + geom_bar(stat='identity') 
```

This didn't really work, because both types are numeric. If we change them to factors, it should work

```r
df<-mtcars
df$gear <- factor(df$gear) # Create a categorical variable
df$cyl <- factor(df$cyl) # Create categorical variable
ggplot(data = df, aes(x=gear, fill=cyl) ) + geom_bar() 
```

```r
ggplot(df, aes(factor(cyl), fill = factor(gear))) +
  geom_bar(position = "dodge2")
```

To draw a pie chart, we start with a bar chart, then convert the geometry to polar_coords:

```r
ggplot(data=mtcars, aes(x=cyl)) + 
  geom_bar() 
```

```r
ggplot(data=mtcars, aes(cyl)) + 
geom_bar() + 
  coord_polar(theta = "y") 
```

That didn't work as expected.  We need to have a single stacked bar chart, rather than multiple bars:

```r
ggplot(data=mtcars, aes(x="", fill = factor(cyl))) + 
  geom_bar() 
```

```r
ggplot(data=mtcars, aes(x = "", fill = factor(cyl))) + 
  geom_bar() + 
  coord_polar(theta = "y") 
```

The rose plot is a variation of the pie chart. For both, we have k data points where each point denotes a frequency or a count. Pie charts and rose plots both use the area of segments of a circle to convey amounts. The pie chart uses a common radius and varies the central angle according to the data.  For the rose plot, the angle is constant (i.e., divide 360 by the number of groups, k) and it is the square root of the radius that is proportional to the data.
https://www.itl.nist.gov/div898/software/dataplot/refman1/auxillar/roseplot.htm

```r
ggplot(data=df,aes(x=cyl,y=mpg,fill=gear))+
        geom_bar(stat="identity")+
        coord_polar()
```
